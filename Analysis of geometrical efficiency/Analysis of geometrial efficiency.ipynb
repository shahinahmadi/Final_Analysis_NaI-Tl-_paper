# Analysis of the geometrical efficiency

## Geometrical efficiency for axial positions of poin source (2 inch)

```python
import pandas as pd
import os
import math
import requests
import numpy as np
from matplotlib import pyplot as plt

data1=pd.read_csv("axial2in_geo.csv")*100*0.98
data1

x_positions=[]
calculation=data1["Calculation"]
simulation =data1["Simulation"]

Positions=np.arange(-2, 2.2, 0.2)
for i in Positions:
    x_positions.append(i)
print(x_positions)

average=0
sigma=0
for number in simulation:
    average+=number/len(simulation)
for number in simulation:
    sigma+=((number-average)*(number-average)/len(simulation))
sigma=math.sqrt(sigma)
e=sigma/math.sqrt(len(simulation))

plt.scatter(x_positions,calculation, color="none", edgecolor="red", label="Calculation")
plt.scatter(x_positions,simulation, color="none", edgecolor="green",marker="^" ,label="Simulation")
plt.errorbar(x_positions,simulation, yerr=e,color="grey", capsize=3,ls="none", label='Simulation uncertanity')
plt.xlabel("Axial position y(cm)")
plt.ylabel("Geometrical efficiency (%)")
plt.rc('legend',fontsize=8)
plt.legend(handlelength=0.5, handleheight=0.2)
leg = plt.legend(loc ="upper right")
plt.legend()
plt.xticks(np.arange(-2,2.2,0.4), fontsize=7)
plt.savefig("Axial_2in_geo.pdf", bbox_inches="tight")
plt.show()
```
