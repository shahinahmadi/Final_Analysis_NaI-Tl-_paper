## Intrinsic efficiency for cylindrical side of 2in for MgO=0.0185cm

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math


data1=pd.read_csv("five.csv")
data11=data1["Calculation"]*100
data12=data1["Simulation"]*100



average=0
sigma=0
for number in data12:
    average+=number/len(data12)
for number in data12:
    sigma+=((number-average)*(number-average)/len(data12))
sigma=math.sqrt(sigma)
e1=sigma/math.sqrt(len(data12))

x_positions=[]

Positions=np.arange(-2, 2.2, 0.2)


for i in Positions:
    x_positions.append(i)
    


plt.scatter(x_positions, data11,s=30, color="none", edgecolor="blue",marker='d', label="Calculation(MgO=0.0185 cm)")
plt.scatter(x_positions, data12,s=30, color="none", edgecolor="cyan",marker='^', label="Calculation(MgO=0.0185 cm)")
plt.legend()
plt.legend(handlelength=0.8, handleheight=0.8)
plt.rc('legend',fontsize=7)
plt.savefig("Cylindrical.pdf", bbox_inches='tight')
plt.show()


data1  


```

## Intrinsic efficiency for cylindrical side of 2in for Al=0.02cm

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math



x_positions=[]


Positions=np.arange(-2,2.2, 0.2)


for i in Positions:
    x_positions.append(i)



final_data=pd.read_csv("ten.csv")


my_data=final_data["Calculation"]*100*0.98
simulation_data=final_data["Simulation"]*100*0.98


average=0
sigma=0
for number in simulation_data:
    average+=number/len(simulation_data)
for number in simulation_data:
    sigma+=((number-average)*(number-average)/len(simulation_data))
sigma=math.sqrt(sigma)
e2=sigma/math.sqrt(len(simulation_data))




plt.scatter(x_positions, data11,s=30, color="none", edgecolor="blue",marker="d", label="Calculation(Mgo=0.0185 cm)")
plt.scatter(x_positions,data12, color='none', edgecolor='cyan', marker='^',label="Simulation(Mgo=0.0185 cm)")
plt.scatter(x_positions, my_data,s=30,color="none", edgecolor="green", label="Calculation(Al=0.02 cm)")
plt.scatter(x_positions, simulation_data,s=30,color="none", edgecolor="red", marker='s',label="Simulation(Al=0.02 cm)")
plt.errorbar(x_positions,data12, yerr=e1,color="black", capsize=3,ls="none", label='Simulation uncertanity(Mgo=0.0185 cm)')
plt.errorbar(x_positions,simulation_data, yerr=e2,color="grey", capsize=3,ls="none", label='Simulation uncertanity(MgO=0.02 cm)')
plt.xlabel("Axial position y(cm)")
plt.ylabel("Intrinsic detection efficiency (%)")
plt.legend()
plt.rc('legend',fontsize=8)
plt.legend(handlelength=0.5, handleheight=0.1)
leg = plt.legend(loc ="best")
plt.ylim(6.5, 13)
plt.xticks(np.arange(-2,2.2, 0.2),fontsize=7)
plt.savefig("Cylindrical.pdf", bbox_inches='tight')
plt.show()
```
