## The intrinsic detection efficiency of circular side of 3in detector for the thickness of 0.0185 cm of Mgo

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math


data1=pd.read_csv("five.csv")
data11=data1["Calculation"]*100*1/0.98
data12=data1["Simulation"]*100*1/0.98



x_positions=[]

Positions=np.arange(-3.6, 3.8, 0.2)


for i in Positions:
    x_positions.append(i)
    
average=0
sigma=0
for number in data12:
    average+=number/len(data12)
for number in data12:
    sigma+=((number-average)*(number-average)/len(data12))
sigma=math.sqrt(sigma)
e1=sigma/math.sqrt(len(data12))

    


plt.scatter(x_positions, data11,s=30, color="none", edgecolor="red", label="Calculation(MgO=0.0185 cm)")
plt.scatter(x_positions,data12, color='none', edgecolor='green', marker='^',label="Simulation")
plt.legend()
plt.legend(handlelength=0.8, handleheight=0.8)
plt.rc('legend',fontsize=7)
#plt.savefig("Circular.pdf", bbox_inches='tight')
plt.show()


data1  

```

## The intrinsic detection efficiency of 3in detector for the thickness of 0.02cm MgO

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math



x_positions=[]


Positions=np.arange(-3.6, 3.8, 0.2)


for i in Positions:
    x_positions.append(i)



final_data=pd.read_csv("ten.csv")
my_data=final_data["Calculation"]*100
simulation_data=final_data["Simulation"]*100


average=0
sigma=0
for number in simulation_data:
    average+=number/len(simulation_data)
for number in simulation_data:
    sigma+=((number-average)*(number-average)/len(simulation_data))
sigma=math.sqrt(sigma)
e2=sigma/math.sqrt(len(simulation_data))



plt.scatter(x_positions, data11,s=30, color="none", edgecolor="blue",marker="d", label="Calculation(MgO=0.0185 cm)")
plt.scatter(x_positions,data12, color='none', edgecolor='cyan', marker='^',label="Simulation(MgO=0.0185 cm)")
plt.scatter(x_positions, my_data,s=30,color="none", edgecolor="green", label="Calculation(MgO=0.02 cm)")
plt.scatter(x_positions, simulation_data,s=30,color="none", edgecolor="red", marker='s',label="Simulation(MgO=0.02 cm)")
plt.errorbar(x_positions,data12, yerr=e1,color="black", capsize=3,ls="none", label='Simulation uncertanity(MgO=0.0185 cm)')
plt.errorbar(x_positions,simulation_data, yerr=e2,color="grey", capsize=3,ls="none", label='Simulation uncertanity(MgO=0.02 cm)')
plt.xlabel("Radial position x(cm)")
plt.ylabel("Intrinsic detection efficiency (%)")
plt.legend()
plt.rc('legend',fontsize=8)
plt.legend(handlelength=0.5, handleheight=0.1)
leg = plt.legend(loc ="best")
plt.ylim(10,16)
plt.xticks(np.arange(-3.6,3.8, 0.4),fontsize=7)
plt.savefig("Circular.pdf", bbox_inches='tight')
plt.show()
```
